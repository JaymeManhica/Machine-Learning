```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df['especies'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df.head()
```

```python
sns.pairplot(iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'especies']], hue='especies')
#
```

# classificando o iris dataset

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)
```

```python
clf = svm.SVC(C=1.0)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
clf.score(X_test, y_test)
```
